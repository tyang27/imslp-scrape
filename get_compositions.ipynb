```python
import pandas as pd
from bs4 import BeautifulSoup

from pprint import pprint
```

```python
!mkdir artist_pages
!mkdir artist_pieces
```

```python
people_links = pd.read_csv('people_id_to_url_map.txt', sep='\s', header=None)
people_links = people_links[1]
people_links
```

```python
for link in people_links[1708:]:
    try:
        link = link.replace('http','https')
        !curl {link} > artist_pages/{link.split(':')[2]}.html
    except:
        print(link)
        continue
```

```python
artist_pages = !ls artist_pages
artist_pages
```

```python
for artist in artist_pages:
    print(artist)
    with open(f'artist_pages/{artist}', 'r') as ifp:
        soup = BeautifulSoup('\n'.join(ifp.readlines()))
        piece_links = (soup.find('table')).find_all('a')
        piece_links = [(x['href'], x['title']) for x in piece_links]
        
        for piece_url, piece_title in piece_links:
            piece_title = piece_title.replace(' ','_').replace('(','').replace(')','').replace(',','')
            artist = artist.replace(' ','_').replace('(','').replace(')','').replace(',','').replace('html','').replace('.','')
            piece_url = 'https://imslp.org' + piece_url
            
            print(piece_url, artist, piece_title)
            !curl "{piece_url}" > artist_pieces/{artist}_{piece_title}.html
            print()
```
